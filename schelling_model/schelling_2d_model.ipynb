# Schelling Segregation Model 2D

Tento agent-based model byl vytvořen Thomasem Schellingem v roce 1971. Model představuje situaci, kdy lidé se segregují do různých skupin podle spokojenosti. V tomto příkladu jsou buňky (lidé) rozděleni na 2 typy. Typ 1 červený a typ 2 modrý. Prázdná místa jsou značena bílou barvou.

### Pravidla

Pravidla pro segregaci jsou velmi jednoduchá

1. Buňka může být prázdná nebo obsahovat 1 nebo 2. --> čísla představují typ lidí
2. Buňka je spokojená pokud procento sousedních buňek je vyšší než nastavený threshold
3. Pokud je buňka (člověk) nespokojená přesouvá se na jakoukoliv volnou buňku v mřížce

### Ukázka simulace

```python
from schelling_class import SchellingModel

model = SchellingModel(size=20, empty_prob=0.05 , type_prob=0.3, threshold=0.3)
model.simulate(max_steps=0, speed=0)
```

## Třída SchellingModel

Třída SchellingModel je inicializována s parametry size, empty_prob, type_prob a threshold. Size je velikost mřížky, empty_prob je pravděpodobnost prázdného pole, type_prob je pravděpodobnost vytvoření typu 1, typ 2 je zbytek buňek, a threshold je threshold pro spokojenost lidí.

### Funkce

1. `initialize_grid()`: inicializuje mřížku s náhodně rozmístěnými buňkami, kterou spolu se součtem buňek typu 1 a 2 vrací. Počet buněk pak slouží k vyreslení grafu spokojenosti
2. `get_neighbors(grid, x, y)`: vrátí sousední buňky buňky na pozici (x, y)
3. `is_satisfied(grid, x, y)`: zjistí, zda buňka (člověk) je spokojená
4. `step(grid)`: přesouvá buňky, které jsou nespokojené na jakoukoliv volnou buňku v mřížce
5. `plot(grid, step_num)`: vykreslí mřížku a počet spokojených lidí na daném kroku simulace
6. `simulate(max_steps, speed)`: simuluje model a vykreslí graf spokojenosti. Parameter max_steps je počet maximalních kroků, které simulace provede k vykreslení grafu. Parameter speed je délka pauzi mezi jednotlivými kroky. Funkce simulace předběžně ukončí pokud jsou všechny buňky v mřížce spokojené.

### Vliv vstupních parametrů na výsledek simulace

1. size --> velikost mřížky ovlivňujě rychlost, variaci a počet kroků simulace
2. empty_prob --> pravděpodobnost prázdného pole ovlivňuje možnost nespokojených buněk se přesouvat. Při menších počet volných míst mohou vzniknout uvězněné nespokojené buňky
3. type_prob --> pravděpodobnost vytvoření typu 1, typ 2 ovlivňuje rychlost simulace, jelikož menší skupiny se rychleji segregují
4. threshold --> threshold ovlivňuje jak moc promíchané skupiny budou a stabilitu skupin
